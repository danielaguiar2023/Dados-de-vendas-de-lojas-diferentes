```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dados das vendas
data = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [420, 250, 380, 210]
}

# Nomes das lojas
lojas = ['A', 'B', 'C', 'D']

# Criação do DataFrame
df_comparacao = pd.DataFrame(data)

# Configuração para gráfico de barras
x = np.arange(len(lojas))  # Posições no eixo x para as lojas
largura = 0.08  # Largura de cada barra

# Criação do gráfico de barras
fig, ax = plt.subplots(figsize=(12, 6))

for i, coluna in enumerate(df_comparacao.columns):
    ax.bar(x + i * largura, df_comparacao[coluna], width=largura, label=coluna)

# Configuração do gráfico
ax.set_title('Comparação de vendas ao longo do ano')
ax.set_xlabel('Lojas')
ax.set_ylabel('Vendas')
ax.set_xticks(x + largura * (len(df_comparacao.columns) - 1) / 2)
ax.set_xticklabels(lojas)  # Define os nomes das lojas como rótulos do eixo X
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Ajusta a posição da legenda
plt.grid(axis='y')  # Adiciona grade horizontal
plt.tight_layout()  # Ajusta o layout para evitar sobreposição

# Exibição do gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dados das vendas
data = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [420, 250, 380, 210]
}

# Nomes das lojas
lojas = ['A', 'B', 'C', 'D']

# Criação do DataFrame
df_comparacao = pd.DataFrame(data, index=lojas)

# Configuração do gráfico
fig, ax = plt.subplots(figsize=(10, 6))

for i, coluna in enumerate(df_comparacao.columns):
    # Ordena as vendas e os nomes das lojas em ordem decrescente
    vendas_ordenadas = df_comparacao[coluna].sort_values(ascending=False)
    lojas_ordenadas = vendas_ordenadas.index

    # Adiciona o gráfico de barras horizontal
    ax.barh(
        y=[f"{coluna} - {loja}" for loja in lojas_ordenadas],
        width=vendas_ordenadas,
        label=coluna,
        alpha=0.8
    )

# Configuração dos rótulos e título
ax.set_title('Vendas organizadas por mês (maiores para menores)', fontsize=14)
ax.set_xlabel('Vendas', fontsize=12)
ax.set_ylabel('Mês - Loja', fontsize=12)
ax.legend(loc='lower right')  # Ajusta a posição da legenda
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Grade apenas no eixo X
plt.tight_layout()  # Ajusta o layout para evitar sobreposição

# Exibição do gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dados das vendas
data = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [420, 250, 380, 210]
}

# Nomes das lojas
lojas = ['A', 'B', 'C', 'D']

# Criação do DataFrame
df_comparacao = pd.DataFrame(data, index=lojas)

# Configuração de estilo profissional
plt.style.use('seaborn-v0_8-muted')
fig, ax = plt.subplots(len(df_comparacao.columns), 1, figsize=(12, 18), sharex=True)

# Cores profissionais
colors = plt.cm.tab20.colors  # Paleta de 20 cores

# Plotando os dados
for i, coluna in enumerate(df_comparacao.columns):
    vendas_ordenadas = df_comparacao[coluna].sort_values(ascending=False)
    lojas_ordenadas = vendas_ordenadas.index
    ax[i].barh(
        y=lojas_ordenadas,
        width=vendas_ordenadas,
        color=colors[i % len(colors)],
        alpha=0.85,
        edgecolor='black'
    )
    ax[i].set_title(f"Vendas em {coluna}", fontsize=14, fontweight='bold')
    ax[i].set_xlabel('Vendas', fontsize=12)
    ax[i].grid(axis='x', linestyle='--', alpha=0.7)
    for j, v in enumerate(vendas_ordenadas):
        ax[i].text(v + 10, j, str(v), va='center', fontsize=10)  # Adiciona o valor nas barras

# Ajustes finais
plt.suptitle('Comparação de Vendas por Mês e Loja', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajusta espaço para o título
plt.show()
```
